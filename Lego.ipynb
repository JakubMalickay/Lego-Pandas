```python
import pandas as pd
lego_sets = pd.read_csv(r'C:\Users\malic\Desktop\pandas\lego_sets.csv')
parent_themes = pd.read_csv(r'C:\Users\malic\Desktop\pandas\parent_themes.csv')
```

```python
import re
```

```python
lego_sets.head()
```

```python
parent_themes.head()
```

```python
df = lego_sets.merge(parent_themes,left_on='parent_theme', right_on='name')
```

```python
df.head()
```

```python
df_licensed = df.loc[df['is_licensed']==True]
```

```python
df_licensed.head(50)
```

```python
df_licensed = df_licensed[df_licensed['set_num'].isnull() == False]
```

```python
star_wars_df = df_licensed.loc[df_licensed['parent_theme'].str.contains('star wars', flags=re.I)]
```

```python
star_wars_count = star_wars_df.shape[0]
```

```python
df_licensed_count = df_licensed.shape[0]
```

```python
the_force = int(star_wars_count/df_licensed_count*100)
```

```python
print(the_force)
```

```python
df_licensed = df_licensed.reset_index()
```

```python
df_licensed = df_licensed.sort_values('year', ascending=True)
```

```python
df_licensed.groupby(['year', 'parent_theme']).head()
```

```python
df_licensed['count'] = 1
```

```python
df_grouped = df_licensed.groupby(['year', 'parent_theme'])['count'].sum().reset_index()
```

```python
df_sorted = df_grouped.sort_values(['count'], ascending = False).drop_duplicates(['year'])
```

```python
df_sorted.sort_values(['year'])
```

```python
new_era = df_sorted.loc[~df_sorted['parent_theme'].str.contains('star wars', flags=re.I)].reset_index()
```
